```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns


#한글 사용
from matplotlib import rc  
rc('font', family='AppleGothic') 			
plt.rcParams['axes.unicode_minus'] = False  

data = pd.read_csv('시간별주차현황.csv') 
df=pd.DataFrame(data)

df.head() #데이터 확인
```

```python
#시간당입구출입=입차대수+출차대수
df['시간당입구출입']=df['입차대수'] + df['출차대수']
df.head(10)
```

```python
#시간대-> 날짜, 시간분리
df['날짜'] = df['시간대'].str.split().str[0]
df['시간'] = df['시간대'].str.split().str[1]
df.head(50)
```

```python
# 주차장명에따라 시간이 같은 행들을 그룹화하고 '시간당입구출입' 값을 합산
# df_group=df.groupby(['주차장명', '시간'])['시간당입구출입'].sum().reset_index()
df_group=df.groupby(['주차장명', '시간'])['시간당입구출입'].mean().reset_index()
# avg=df_group['시간당입구출입']/214
# avg
df_group
```

```python
#주차장별 시간당 입구출입 시각화
#공통적으로 주차장 입구 출입이 가장 많은 시간대 확인

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_group, x='시간', y='시간당입구출입', hue='주차장명', marker='o', palette='Set2')
plt.title('주차장별 시간당 입구 출입')
plt.xlabel('시간')
plt.ylabel('시간당 입구 출입')
plt.xticks(rotation=45)
plt.legend(title='주차장명', loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3)
plt.tight_layout()
plt.show()
#전체적으로 시간당 입구 출입이 많은 시간대는 아침시간: 08, 09, 10  저녁시간: 18, 19, 20시 
```

```python
#주차장별 아침/ 저녁 데이터
#가장 출입이 많은 시간대 시간당 입구 출입 합산 및 평균
morning = df_group[df_group['시간'].isin(['08', '09', '10'])]  #주차장별 아침 데이터 (08, 09, 10시)
evening =df_group[df_group['시간'].isin(['18', '19', '20'])]   #주차장별 저녁 데이터 (18, 19, 20시)
morning_avg=morning.groupby('주차장명')['시간당입구출입'].mean()   #주차장별 아침 데이터 평균
evening_avg=evening.groupby('주차장명')['시간당입구출입'].mean()   #주차장별 저녁 데이터 평균

morning_avg #데이터 확인
evening_avg #데이터 확인
```

```python
# 아침/저녁 평균 시각화(데이터 파악)
plt.figure(figsize=(10, 6))

bar_width = 0.35
index = range(len(morning_avg))

plt.bar(index, morning_avg, bar_width, label='아침')
plt.bar([i + bar_width for i in index], evening_avg, bar_width, label='저녁')

plt.ylabel('평균 시간당 입구 출입')
plt.title('주차장별 아침과 저녁 시간대 평균 출입 비교')
plt.xticks([i + bar_width / 2 for i in index], morning_avg.index, rotation='vertical')

plt.legend()
plt.tight_layout()
plt.show()

#지역에 따라 아침과 저녁 입구혼잡도의 차이가 있다.
#저녁에 입구혼잡도가 높은 주차장이 근소하게 더 많다.
```

```python
#데이터 분석 결과
#주차장별 혼잡도
#아침 평균 혼잡도+저녁 평균 혼잡도의 평균
All=df_group[df_group['시간'].isin(['08', '09', '10','18', '19', '20'])]
df_result=All.groupby('주차장명')['시간당입구출입'].mean() 

df_result  #최종 데이터 확인
```

```python
#최종 데이터 시각화
#가장 입구혼잡도가 높은 주차장은 어디일까?
# 시각화
plt.figure(figsize=(10, 6))
df_result.plot(kind='bar')
plt.xlabel('주차장명')
plt.ylabel('평균 시간당 입구 출입')
plt.title('주차장별평균혼잡도')
plt.xticks(rotation='vertical')
plt.tight_layout()
plt.show()

#결과: 천호역이 입구혼잡도가 가장 높음, 신월 4동이 입구혼잡도가 가장 낮음
```

```python
# df_result CSV 파일로 저장
df_result.to_csv('ParkingTraffic.csv', header=True)
```
